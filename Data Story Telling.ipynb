```python
import pandas as pd
import numpy as np
from plotnine import *
```

```python
df = pd.read_csv('adult.csv')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
df['educational-num'].unique()
```

```python
df['occupation'].unique()
```

```python
# Is there a correlation between education years and higher earnings
ggplot(df, aes(x='educational-num', fill = 'income')) + geom_bar(position = 'fill') + theme(axis_text_x = element_text(angle=90, hjust =1)) + ylab('Ratio') + xlab('Educational Years')

```

```python
# There is a strong trend for the ratio of higher earners and the education years. The longer a person is in
# school the higher the likelihood that they will earn more.
```

```python
dfover = df[df['income'] == '>50K']
```

```python
dfover.shape
```

```python
# What is the marital status of higher earners
ggplot(dfover, aes(x='marital-status')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# For those earning higher than 50K a strong majority is married.
```

```python
# What type of profession are higher earners doing?
ggplot(dfover, aes(x='occupation')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# Higher earners are mostly in Exec-managerial roles or a prof-specialty of sorts.
```

```python
# How many hours are higher earnings working in a week
ggplot(dfover, aes(x='hours-per-week')) + geom_histogram(bins=10) + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# Higher earners are working mostly a little under 50 hours a week
```

```python
# How old are the higher earners
ggplot(dfover, aes(x='age')) + geom_histogram(bins=15) + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
dfunder = df[df['income'] == '<=50K']
```

```python
dfunder.shape
```

```python
# What is the marital status of lower earners?
ggplot(dfunder, aes(x='marital-status')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# People earning less than 50K have a higher number of them that are unmarried. There is also a higher number of 
# people that are Divorced
```

```python
# What are lower earners tyoically occupied as?
ggplot(dfunder, aes(x='occupation')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# The occupations for lower earners are much more spread out than higher earners.
```

```python
# What is the race and gender for higher earners?
ggplot(dfover, aes(x='race', fill = 'gender')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1)) + coord_flip()
```

```python
#Most of the high earners are white males
```

```python
#How many hours are lower earners working a week
ggplot(dfunder, aes(x='hours-per-week')) + geom_histogram(bins=10) + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# Lower earners are working a little less than 50 hours a week. This is not much different than higher earners.
```

```python
# What is the age of lower earners?
ggplot(dfunder, aes(x='age')) + geom_histogram(bins=15) + theme(axis_text_x = element_text(angle=90, hjust =1))
```

```python
# lower earners seem to be younger than higher earners. There is a slight skew towards the younger ages
```

```python
# What is the race and gender of lower earners?
ggplot(dfunder, aes(x='race', fill = 'gender')) + geom_bar() + theme(axis_text_x = element_text(angle=90, hjust =1)) + coord_flip()
```

```python
# Most lower earners are white males. Although the ratio of men to woman is less than that of higher earners. There are
# also more black people as lower earners
```

```python
#What is ratio of low vs high earners by race
ggplot(df, aes(x='race', fill = 'income')) + geom_bar(position= 'dodge') + theme(axis_text_x = element_text(angle=90, hjust =1)) + coord_flip()
```

```python
# White has the best ratio for higher earners to lower earners. It is about 1/4th of the of white earning above 50K
```
